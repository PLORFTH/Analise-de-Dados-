```python
#LOCALIZADOR: variavel.loc[rotulo] "Onde o rótulo é indice a ser localizado"
series_dados = pd.Series(lista_nomes, index=cpfs)
#atribuimos os dados de uma tabela inteira a uma variavel para podermos manipular a tabela sem mecher nos dados reais.
series_dados.loc['111.111.111-11']
#nessa linha de comando, nós implementamos um código para saber quem é o dono do CPF (x), que no nosso caso foi o "Howard".
#é um localizador do pandas, onde nós podemos buscar na tabela qual "objeto" é o dono do dado(a) e qual "objeto" é o dono do dado(b)
```

```python
#Tudo isso é para saber como tirar informações de uma Serie ou Tabela.

from pandas.core import series
series_dados = pd.Series([10.2, -1, None, 15, 23.4])
print('Quantidade de linhas = ', series_dados.shape)#retorna uma tupla com o numero de linhas
print('Tipo de dados ', series_dados.dtypes)# retorna o tipo de dados, se for misto será "object"
print('Os valores são únicos? ', series_dados.is_unique)#Verifica se os valores são únicos(sem duplicações [redundancias])
print('Existem valores nulos? ', series_dados.hasnans)#verifica se existe valor nulo.
print('Quantos valores existem? ', series_dados.count())#conta quantos valores existem (exclui os núlos).

print('Qual o menor valor? ', series_dados.min())#Extrai o dado com menor valor da Serie (os dados precisam ser do mesmo tipo)
print('Qual o maior valor? ', series_dados.max())#Extrai o valor máximo, com a mesma condição do mínimo.
print('Qual a média aritimética? ', series_dados.mean())#Extrai a média aritimética de uma Series numérica.
print('Qual o desvio padrão? ', series_dados.std())#Extrai o desvio padrão de uma série numérica.
print('Qual a mediana? ', series_dados.median())#Extrai a mediana de uma Serie numérica.

print('\nResumo:\n', series_dados.describe())#Exibe um resumo sobre os dados na Serie
```

```python
#SOBRE DATAFRAME
#metodo construtor: pandas.DataFrame(data=None, index=None, colums=None, dtype=None, copy=False)
```

```python
lista_nomes = 'Howard Ian Peter Jonah Kellie'.split()
lista_cpfs = '111.111.111-11 222.222.222-22 333.333.333-33 444.444.444-44 555.555.555-55'.split()
lista_emails = 'Howard_dada@barato.com Ian_donate@oferta.com Peter_barganha@compre.com Jonah_vendas@caduca.com Kellie_metas@modelagem.com'.split()
lista_idades= [32,22,25,29,38]
pd.DataFrame(lista_nomes, columns=['Nome'])#aqui criamos uma tabela com os nomes que estavam na lista
```

```python
pd.DataFrame(lista_nomes, columns=['Nome'], index=lista_cpfs)#criamos a tabela com cada um com o seu CPF
```

```python
#o comando zip() cria tuplas, cada uma composta por um valor de cada lista
dados = list(zip(lista_nomes, lista_cpfs, lista_idades, lista_emails))#Guardamos os dados em uma só lista, (isso é uma lista de tuplas)
pd.DataFrame(dados, columns=['NOME','CPF', 'IDADE','EMAIL'])#criamos uma tabela a partir da lista que nós criamos
```

```python
import pandas as pd

#Construir tabela a partir de um DICIONARIO
#Cad  chave deve estar associada a uma lista de mesmo TAMANHO

dados= {
    'nomes': 'Howard Ian Peter Jonah Kellie'.split(),
    'cpfs': '111.111.111-11 222.222.222-22 333.333.333-33 444.444.444-44 555.555.555-55'.split(),
    'emails': 'Howard_dada@barato.com Ian_donate@oferta.com Peter_barganha@compre.com Jonah_vendas@caduca.com Kellie_metas@modelagem.com'.split(),
    'idades': [32, 22, 25, 29, 38]
}
# um dicionario vocÊ pode dar todos os dados de uma só vez, e o DataFrame faz tudo sozinho gerando a tabela da forma correta
pd.DataFrame(dados)
```

```python
#Como extrair informações de um DataFrame
df_dados = pd.DataFrame(dados)
print('\nInformações do DataFrame: \n')
print(df_dados.info()) #Apresenta informações sobre a estrutura do DF

print('\nQuantidade de linhas e colunas = \n', df_dados.shape)#retorna uma tupla com o número de linhas e colunas.
print('\nTipo de dados:\n',df_dados.dtypes)#Retorna o tipo de dados, para cada coluna, se for misto será "object".

print('\nQual o menor valor de cada coluna? \n',df_dados.min())#Extrai o menor de cada coluna.
print('\nQual o maior valor de cada coluna? \n',df_dados.max())#Extrai o maior valor de cada coluna.
print('\nQual a média aritimética? \n',df_dados.mean())#Extrai a média aritimética de cada coluna numérica.
print('\nQual o desvio padrão? \n',df_dados.std())#Extrai o desvio padrão de cada coluna numérica.
print('\nQual a mediana? \n',df_dados.median())#Extrai a mediana de cada coluna numérica.

print('\nResumo: \n',df_dados.describe())#Exibe um resumo

df_dados.head()#Exibe as 5 primeiros registros do DataFrame.
```
